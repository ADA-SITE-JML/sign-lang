```python
import os
import cv2
```

```python
src_parent = "../data/"
partitions = ["train", "val", "test"]

for dir_name in os.listdir(src_parent):
    
    if dir_name in partitions:
        src_path = os.path.join(src_parent, dir_name)
        
        
        for gloss in os.listdir(src_path):
            gloss_path = os.path.join(src_path, gloss)
            
            for i, video in enumerate(os.listdir(gloss_path)):
                video_path = os.path.join(gloss_path, video)
                
                cap = cv2.VideoCapture(video_path)
                success ,image = cap.read()
                count = 0
                dest_path = os.path.join(src_parent, "torch-video-dataset", dir_name, gloss, f"{i:05d}")
                if not os.path.exists(dest_path):
                    os.makedirs(dest_path)
                    
                while success:
                    file_name = f"image_{count:05d}.jpg"
                    file_path = os.path.join(dest_path, file_name)
                    cv2.imwrite(file_path, image)   
                    success, image = cap.read()
                    print('Read a new frame: ', success)
                    count += 1
```
